## 使用Dask-ml分類

```python
from dask_ml.cluster import KMeans
import dask_ml.datasets #pip install dask-ml
```

## 以KMeans做

```python
X, y = dask_ml.datasets.make_blobs(n_samples = 500000, random_state = 0, centers = 5, chunks = 10)  #在dataset中取500000筆資料,一塊為50000筆資料, random state = 0, 分成五種資料
```

```python
X
```

```python
#請KMeans分成五類
#開始訓練
km = dask_ml.cluster.KMeans(n_clusters = 5, init_max_iter = 2, oversampling_factor = 10)
km.fit(X)
```

```python
fig, ax = plt.subplots()
ax.scatter(X[::1000, 0], X[::1000, 1], marker = '.', c = km.labels_[::1000],
           cmap = 'viridis', alpha = 0.25)
```
